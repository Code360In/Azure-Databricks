# Day29 - Performance Tuning

## Caching

### Get some sample Data

```
DF1 = spark.read.format("CSV").option("inferSchema", "true").option("header","true").load("/databricks-datasets/COVID/covid-19-data/us-states.csv")
```

### Aggregate some data

```
DF2 = DF1.groupby("state").count().collect()
DF3 = DF1.groupby("date").count().collect()
DF4 = DF1.groupby("cases").count().collect()
```

it was running on my cluster almost 2 seconds

### Cache data

```
DF1.cache()
DF1.count()
```

```
DF2 = DF1.groupby("state").count().collect()
DF3 = DF1.groupby("date").count().collect()
DF4 = DF1.groupby("cases").count().collect()
```

When cached it was running on my cluster under 1 second
